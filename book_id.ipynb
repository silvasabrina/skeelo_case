```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

delivered_books = pd.read_csv("delivered_books.csv", names=["delivered_at","book_id"])
```

```python
book_ids = delivered_books["book_id"].value_counts().rename_axis('book_id').reset_index(name='counts') 

sorted_books = book_ids.sort_values(by=['counts'], ascending=True)  # dataframe ordenado pela coluna count de forma crescente

xpos = sorted_books["book_id"]   # eixo x com valores dos ids dos livros
ypos = sorted_books["counts"]    # eixo y com valores da frequência da distribuição dos livros

xpos = np.arange(len(sorted_books["book_id"]))  # criação do array que fará o mapeamento dos valores dos xticks
plt.xticks(xpos, sorted_books["book_id"].array) # alteração dos valores dos xticks com os ids dos livros

# configurações básicas e plot do gráfico
plt.stem(xpos,ypos)
plt.title("Quantidade de livros distribuídos por id")
plt.ylabel("Quantidade de livros distribuídos")
plt.xlabel("Id")
plt.show()
```
