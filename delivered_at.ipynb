```python
months = ["Jan", "Fev", "Mar", "Abr", "Mai", "Jun", "Jul", "Ago", "Set", "Out", "Nov", "Dez"]
xpos = np.arange(len(months))
ypos_21 = [
    quantity_delivered_at_jan_2021,
    quantity_delivered_at_fev_2021,
    quantity_delivered_at_mar_2021,
    quantity_delivered_at_abr_2021,
    quantity_delivered_at_mai_2021,
    quantity_delivered_at_jun_2021,
    quantity_delivered_at_jul_2021,
    quantity_delivered_at_ago_2021,
    quantity_delivered_at_set_2021,
    quantity_delivered_at_out_2021,
    quantity_delivered_at_nov_2021,
    quantity_delivered_at_dez_2021
]
ypos_22 = [
    quantity_delivered_at_jan_2022,
    quantity_delivered_at_fev_2022,
    quantity_delivered_at_mar_2022,
    quantity_delivered_at_abr_2022,
    quantity_delivered_at_mai_2022,
    quantity_delivered_at_jun_2022,
    quantity_delivered_at_jul_2022,
    quantity_delivered_at_ago_2022,
    quantity_delivered_at_set_2022,
    quantity_delivered_at_out_2022, # outlier 
    quantity_delivered_at_nov_2022,
    quantity_delivered_at_dez_2022
]
```

```python
# configurações básicas e plot dos gráficos
plt.xticks(xpos, months)
plt.title("Quantidade de livros distribuídos por mês")
plt.ylabel("Quantidade de livros distribuídos")

plt.bar(xpos+0.2, ypos_21, width=0.4, color="#7743DB", label="2021")
plt.bar(xpos-0.2, ypos_22, width=0.4, color="#3B3486", label="2022")
plt.legend()
```
